```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
n = 100
phi = 0.5  
theta = 0.3  

# Generate ARMA(1,1) process
errors = np.random.normal(size=n)
data = np.zeros(n)
data[0] = errors[0]
for i in range(1, n):
    data[i] = phi * data[i - 1] + errors[i] + theta * errors[i - 1]

# Innovation algorithm
residuals = np.zeros(n)
for i in range(1, n):
    sum_term = 0.0
    for j in range(1, i + 1):
        sum_term += phi ** j * errors[i - j]
    residuals[i] = errors[i] - theta * errors[i - 1] - sum_term
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot original time series data
plt.subplot(2, 1, 1)
plt.plot(data, label='Original Data')
plt.title('Original Time Series Data')
plt.legend()

# Plot residuals obtained from the innovation algorithm
plt.subplot(2, 1, 2)
plt.plot(residuals, label='Residuals')
plt.title('Residuals (Innovations) from ARMA(1,1) Model')
plt.legend()

plt.tight_layout()
plt.show()
```
